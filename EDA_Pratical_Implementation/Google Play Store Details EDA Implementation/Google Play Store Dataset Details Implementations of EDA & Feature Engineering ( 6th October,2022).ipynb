```python
df_copy = df.copy()
```

```python
df_copy = df_copy.drop(df_copy.index[10472])   ## There we are droping the index = 10472 from a dataset
```

```python
df_copy['Reviews'] = df_copy['Reviews'].astype('int')  ## Here we are Converting Dtpye from Object to 'Int'
```

```python
for i in df_copy['Size'].str:
    print(i)
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('M','000')   ## Here we are using Replace function
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('k','')    ## Here we are using Replace Function
```

```python
df_copy['Size'].unique()
```

```python
import numpy as np
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('Varies with device',str(np.nan))   ## Here we are Replacing with Nan value.
```

```python
df_copy['Size'].dtype
```

```python
df_copy['Size'] = df_copy['Size'].astype('float')
```

```python
df_copy['Size'].dtype
```

```python
df_copy['Size'].head()
```

```python
df_copy['Size'][2]*1000
```

```python
for i in df_copy['Size']:
    if i < 10 :
        df_copy['Size']= df_copy['Size'].replace(i, i*1000)      ## Here we are using Scaling Function
```

```python
df_copy['Size'].head()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'] = df_copy['Size']/1000    ## Here we are using Scaling Function
```

```python
df_copy['Installs']
```

```python
df_copy['Installs'].dtype
```

```python
df_copy['Installs'].unique()
```

```python
chars_to_remove = ['+',',','$']
cols_to_clean = ['Installs', 'Price']

for item in chars_to_remove :
    for col in cols_to_clean:
        df_copy[col] = df_copy[col].str.replace(item,'')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs'] = df_copy['Installs'].astype('int')
```

```python
df_copy['Price'] = df_copy['Price'].astype('float')
```

```python
df_copy.head(3)
```

```python
df_copy['Last Updated']
```

```python
df_copy['Last Updated'].dtype
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
```

```python
df_copy['Last Updated'].dtype
```

```python
df_copy['Day'] = df_copy['Last Updated'].dt.day
```

```python
df_copy['Month'] = df_copy['Last Updated'].dt.month
```

```python
df_copy['Year'] = df_copy['Last Updated'].dt.year
```
